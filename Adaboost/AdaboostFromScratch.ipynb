```python
def error_rate(ratio):
    alpha  = 0.5*np.log((1-ratio)/ratio)
    return alpha
```

```python
def find_error(dataFrame):
    count = 0
    for i in range(dataFrame.shape[0]):
        if dataFrame['y_pred'].iloc[i] == dataFrame['label'].iloc[i]:
            count += 1
    return (dataFrame.shape[0] - count) / dataFrame.shape[0]


def weight_class(dataFrame):
    new_weights = []
    
    error = find_error(dataFrame)
    error = np.clip(error, 1e-10, 1 - 1e-10)
    alpha = error_rate(error)
    
    for i in range(dataFrame.shape[0]):
        if dataFrame['label'].iloc[i] == dataFrame['y_pred'].iloc[i]:
            w = dataFrame['weights'].iloc[i] * np.exp(-alpha)
        else:
            w = dataFrame['weights'].iloc[i] * np.exp(alpha)
        
        new_weights.append(w)
    
    new_weights = np.array(new_weights)
    new_weights = new_weights / np.sum(new_weights)
    
    return new_weights
```

```python
df['updated_weights'] = weight_class(df)
```

```python
df
```

```python
df['nomalized_weights'] = df['updated_weights']/df['updated_weights'].sum()
```

```python
df['cumsum_upper'] = np.cumsum(df['nomalized_weights'])
df['cumsum_lower'] = df['cumsum_upper'] - df['nomalized_weights']
```

```python
def new_dataset(df):
  indices = []

  for i in range(df.shape[0]):
    a = np.random.random()
    for index,row in df.iterrows():
      if row['cumsum_upper'] > a and a > row['cumsum_lower']:
        indices.append(index)
  return indices
     
idx_values = new_dataset(df)
idx_values
     
```
